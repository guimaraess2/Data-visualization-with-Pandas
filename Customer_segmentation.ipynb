Importação das bibliotecas

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import cufflinks as cf
 
```

Carregar o dataset

```
df = pd.read_csv('Mall_Customers.csv')
```

visualizar as primeiras 5 linhas

```
df.head()
```

Verificar o tipo de dados

```
df.dtypes
```

Informações estatísticas

```
df.describe()
```

Qual é a quantidade dados existentes, por gênero,  na base de dados? 

```
sns.catplot(x='Genre', kind='count' , palette='Set2',data=df)
```

Qual a faixa de idade ideal para direcionar a venda de produtos?

```
sns.distplot(df['Age'], color='green', kde=True, bins=8)
plt.title('Distribuição')
```

Qual a correlação entre os atributos idade, renda anual e pontuação por gastos?

```
correlacao=df[['Age','Spending Score (1-100)','Annual Income (k$)']].corr()
```

```
sns.heatmap(correlacao, annot=True, cmap='PiYG')
```

Qual a concentração da renda anual por idade?

```
sns.scatterplot(x='Age' , y='Annual Income (k$)' , color='blue', hue='Genre',data=df)
```

Visualizando melhor a concentração dos dados

```
sns.catplot(x='Genre', y='Spending Score (1-100)', kind='violin', data=df)
```

Qual gênero possui maior amplitude de gastos?

```
sns.boxplot(x='Genre', y='Spending Score (1-100)', data=df)
```

Em que faixa de idade se concentra a renda?

Em que idade se concentra maior quantidade de pontuação por gastos?

```
sns.jointplot(x='Age', y='Spending Score (1-100)', kind='hex', color='b', data=df, height=4)
```

Quais as relações entre todas as variáveis da base de dados?

```
sns.pairplot(df, palette='Set1', hue='Genre');
```
